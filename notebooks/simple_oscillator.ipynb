```
import numpy as np
import matplotlib.pyplot as plt

m = 1.0  # mass
k = 1.0  # spring constant
omega = np.sqrt(k/m)

t = np.linspace(0, 20, 500)
A = 1.0
x = A * np.cos(omega * t)

plt.plot(t, x)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Simple Harmonic Motion')
plt.show()
```
